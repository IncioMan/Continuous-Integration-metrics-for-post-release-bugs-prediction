```python
pr_files = glob.glob(f"{folder}/*.pkl")
prs_raw = []
for pr_file in pr_files:
    with open(pr_file, "rb") as f:
        prs_raw = prs_raw + pickle.load(f)
```

```python
pp.pprint(prs_raw[0][0])
```

```python
prs_commits_data = []
prs_data = []
for pr in prs_raw:
    if not "number" in pr[0]:
        continue
    prs_data.append((pr[0]["number"], pr[0]["title"], pr[0]["base"]["sha"], pr[0]["head"]["sha"]))
    for commit in pr[1]:
        prs_commits_data.append((pr[0]["number"], commit["sha"]))
```

```python
pr_commits = pd.DataFrame(prs_commits_data, columns=["pr_number", "commitsha"])
pr_commits
```

```python
prs = pd.DataFrame(prs_data, columns=["pr_number", "title", "base_sha", "head_sha"])
prs
```

```python
pr_commits.to_csv(f"{csv_folder}/pr_commits.csv")
prs.to_csv(f"{csv_folder}/prs.csv")
```
